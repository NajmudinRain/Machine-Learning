```python
import seaborn as sns
import pandas as pd
```

```python
tips = sns.load_dataset('tips')
```

```python
# tips.sample(5)
tips.head()
```

```python
titanic = pd.read_csv('Titanic-dataset.csv')
titanic.head()
```

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
# Scaterplot- (Numerical - Numerical)
```

```python
sns.scatterplot(x = tips['total_bill'], y = tips['tip'])
```

```python
sns.scatterplot(x = tips['total_bill'], y = tips['tip'], hue = tips['sex'], size = tips['smoker'], style = tips['time'])
```

BarPlot - (numerical -categorical)

```python
sns.barplot(x = titanic['Pclass'], y = titanic['Fare'], hue = titanic['Sex'])
```

BoxPlot- (numerical - categorical)

```python
sns.boxplot(x=titanic['Sex'], y=titanic['Age'], hue= titanic['Survived'])
```

DistPlot - (numerical - categorical)

```python
sns.displot(titanic[titanic['Survived']==0]['Age'])
sns.displot(titanic[titanic['Survived']==1]['Age'])
```

```python
#HEATMAP(categorical - categorical)
```

```python

# What pd.crosstab() Does:
# It creates a cross-tabulation table  a contingency table that shows the frequency of combinations between two (or more) categorical variables.
pd.crosstab(titanic['Pclass'], titanic['Survived'])
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'], titanic['Survived']))
```

```python
titanic.groupby("Sex")["Survived"].mean() * 100
```
