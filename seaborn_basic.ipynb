```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load example dataset
df = sns.load_dataset("tips")
```

```python
sns.histplot(df['total_bill'], bins=20, kde=False)
plt.title("Histogram of Total Bill")
plt.xlabel("Total Bill")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.kdeplot(df['total_bill'], fill=True)
plt.title("KDE of Total Bill")
plt.xlabel("Total Bill")
plt.show()
```

```python
sns.boxplot(x='day', y='total_bill', data=df)
plt.title("Box Plot of Total Bill by Day")
plt.show()
```

```python
sns.scatterplot(x='total_bill', y='tip', data=df)
plt.title("Scatter Plot: Total Bill vs Tip")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

```python
sns.barplot(x='day', y='total_bill', data=df, estimator=sum)
plt.title("Total Bill Sum by Day (Bar Plot)")
plt.show()
```

```python
sns.countplot(x='day', data=df)
plt.title("Count of Records per Day")
plt.ylabel("Count")
plt.show()
```
